```python
import pandas as pd

file_path = "./car details v4.csv"
cars_data = pd.read_csv(file_path)
cars_data = cars_data.drop(['Max Power','Max Torque'],axis=1)
cars_data.head(20)
```

```python
missing_values = cars_data.isna().sum()
missing_types = cars_data.dtypes
missing_values,missing_types
```

```python
cars_data.Engine = cars_data['Engine'].fillna(cars_data['Engine'].mode()[0])
cars_data.Drivetrain = cars_data['Drivetrain'].fillna(cars_data['Drivetrain'].mode()[0])
```
